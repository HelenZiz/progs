# Центральная предельная теорема
## Равномерное распределение
### Будем расматривать равномерное распределение на отрезке [0, 1]
### Его математическое ожидание равно E[X] = 0.5
### Дисперсия D[X] = 1/12

```python
from scipy.stats import uniform
import matplotlib.pyplot as plt
import numpy as np
import math
import scipy.stats as sts
fig, ax = plt.subplots(1, 1)

mean, var, skew, kurt = uniform.stats(moments='mvsk')

x = np.linspace(uniform.ppf(0.01),uniform.ppf(0.99), 100) #задаем массив из 100 элементов от 0 до 1
ax.plot(x, uniform.pdf(x),'r-', lw=5, alpha=0.6, label='theory') #строим теоретическую плотность распределения

vals = uniform.ppf([0.001, 0.5, 0.999])
np.allclose([0.001, 0.5, 0.999], uniform.cdf(vals))
r = uniform.rvs(size=1000) #генерируем случайную выборку из 1000 значений с равномерным распределением

ax.hist(r, density=True, histtype='stepfilled', alpha=0.2) #строим гистограмму
ax.legend(loc='best', frameon=False)
plt.show()
```

## Распределение выборочного среднего случайной величины при разных объёмах выборок:
### n = 10

```python
import math
import scipy
n=10
sample10=np.array([]) #обозначим массив выборочных средних
i=1
while i <= 1000: #цикл генерации случайных выборок количества n из функции
    sample10i = np.random.uniform (0,1,n) #генерация выборки
    mean10i=np.mean(sample10i) #определение выборочного среднего (в.с.)
    sample10=np.append(sample10,mean10i) #добавляем в массив новое значение в.с.
    i+=1
#определим параметры нормального распределения описывающие массив 1000 в.с. при n=10
E=0.5
D=(1./12) #дисперсия равномерного распределения
D10=D/n #дисперсия для нормального распределения приближающего выборку средних
sigma10=math.sqrt(D10)#сигма для нормального распределения выборки средних
norm10 = sts.norm(E, sigma10)#определяем нормальное распределение с расчетными характеристиками равномерного
x = np.linspace(0,1,1000)
theory10 = norm10.pdf(x[:1000])#получаем плотность распределения
#строим гистограмму массива средних при n=10 и плотность описывающую этот массив нормального распределения
fig, ax = plt.subplots(figsize=[12, 8])
plt.plot(x[:1000], theory10, label='theory n=10')
plt.hist(sample10, bins =80)
plt.ylabel('number of samples')
plt.xlabel('$x$')
plt.legend(loc='best')
plt.show()
```
